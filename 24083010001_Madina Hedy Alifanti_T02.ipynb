```python
import math
```

```python
#1
def no1(n):
    kiri = sum(2*i - 1 for i in range(1, n+1))
    kanan = n**2
    hasil = kiri == kanan
    return f"Jumlah bilangan ganjil pertama sampai {n}: {kiri}, {n}^2 = {kanan} → {hasil}"

n1 = int(input("Masukkan angka: "))
print(no1(n1))
```

```python
def no2(n):
    kiri = 1 + sum(2**i for i in range(1, n+1))
    kanan = 2**(n+1) - 1
    hasil = kiri == kanan
    return f"Jumlah 1 + 2 + ... + 2^{n} = {kiri}, 2^({n+1}) - 1 = {kanan} → {hasil}"

n2 = int(input("Masukkan angka: "))
print(no2(n2))
```

```python
def no3(n):
    kiri = sum(i**2 for i in range(1, n+1))
    kanan = n*(n+1)*(2*n+1)//6
    hasil = kiri == kanan
    return f"Jumlah kuadrat 1^2 + ... + {n}^2 = {kiri}, rumus = {kanan} → {hasil}"

n3 = int(input("Masukkan angka: "))
print(no3(n3))
```

```python
def no4(n):
    kiri = math.factorial(n)
    kanan = 2**(n-1)
    hasil = kiri >= kanan
    return f"{n}! = {kiri}, 2^({n}-1) = {kanan} → {hasil}"

n4 = int(input("Masukkan angka: "))
print(no4(n4))
```

```python
def no5(n):
    kiri = 2**n
    kanan = n**2
    hasil = kiri > kanan
    return f"2^{n} = {kiri}, {n}^2 = {kanan} → {hasil}"

n5 = int(input("Masukkan angka: "))
print( no5(n5))
```
