```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
abstrak = "Komunikasi dan membangun hubungan kini dipermudah dengan  menggunakan teknologi seperti aplikasi media sosial. Salah satu pelaku yang berkontribusi terhadap kesuksesan platform digital marketing seperti media sosial adalah penggunaan influencer. Pemasaran yang efektif melibatkan dalam menjaga produk di garis  depan pada pikiran konsumen sehingga mereka memilih untuk membeli produk yang ditawarkan. Tujuan dari penelitian ini adalah untuk mengetahui bagaimana digital marketing menggunakan sosial media influencer yang dapat mempengaruhi perilaku pembelian konsumen. Menurut aliran AISAS yakni micro-influencer, macroinfluencer, dan mega-influencer dapat membuat konten yang menarik sehingga mempengaruhi minat banyak konsumen. Jumlah suka dan komentar di akun influencer dapat menunjukkan minat follower terhadap akun tersebut.  Influencer juga dapat mengkomunikasikan berbagai jenis promosi melalui gaya visual dan bahasa pribadi mereka. Ini ditunjukkan dengan menggunakan fotografi dan copywriting yang menarik. Potensi netizen yang terkena dampak untuk mencari produk (search), membeli (action), dan review produk sudah jelas, yang dapat berdampak pada perusahaan yang dapat mempromosikan produknya kepada influencer. Statistik dan tren Instagram penting untuk tahun 2022 yakni pengiklan dapat mencapai 1,440 miliar pengguna di Instagram pada Juli 2022, menempatkannya di peringkat ke-4 platform media sosial paling 'aktif' di dunia. Instagram membatasi penggunaan platformnya untuk orang berusia 13 tahun ke atas, sehingga sangat membantu untuk mengetahui bahwa 45,8 persen audiens yang memenuhi syarat di Indonesia menggunakan Instagram pada tahun 2022. Akibatnya, influencer menjadi salah satu strategi pemasaran yang populer bagi para influencer karena mereka dapat menyajikan postingan yang dapat mempromosikan citra dan karakteristik brand, yang berdampak signifikan bagi konsumennya."
kalimat_abstrak = re.split(r'[.!?]\s*', abstrak)
kalimat_abstrak = [k for k in kalimat_abstrak if k]

for i, kalimat in enumerate(kalimat_abstrak, start=1):
    kata_abstrak = kalimat.split()
    print(f"\nKalimat {i}: {kalimat.strip()}")
    print("Kata:", kata_abstrak)
```

```python
vokal = "aiueoAIUEO"
data = []

# Loop semua kata di setiap kalimat
for kalimat in kalimat_abstrak:
    kata_abstrak = kalimat.split()
    for kata in kata_abstrak:
        jumlah_vokal = sum(1 for huruf in kata if huruf in vokal)
        jumlah_konsonan = sum(1 for huruf in kata if huruf.isalpha() and huruf not in vokal)
        data.append({
            "Kata": kata,
            "Vokal": jumlah_vokal,
            "Konsonan": jumlah_konsonan
        })

# Buat DataFrame
df_kata_abstrak = pd.DataFrame(data)

# Tampilkan hasil
df_kata_abstrak
```

```python
# --- Pisahkan kalimat dari teks abstrak ---
kalimat_abstrak = re.split(r'[.!?]\s*', abstrak)
kalimat_abstrak = [k.strip() for k in kalimat_abstrak if k]  # hapus kosong & spasi

# --- Hitung jumlah kata per kalimat ---
data_kalimat = []
for i, kalimat in enumerate(kalimat_abstrak, start=1):
    jumlah_kata = len(kalimat.split())
    data_kalimat.append({
        "Kalimat_ke": i,
        "Kalimat": kalimat,
        "Jumlah_Kata": jumlah_kata
    })

# --- Buat DataFrame ---
df_kalimat_abstrak = pd.DataFrame(data_kalimat)

# --- Tampilkan hasil ---
df_kalimat_abstrak
```

```python
# --- PIVOT 2: Cacah kata unik dalam tiap kombinasi ---
# Karena setiap baris di df sudah 1 kata, hasilnya sama dengan frekuensi kombinasi
pivot_cacah = df_kata_abstrak.groupby(['Vokal', 'Konsonan']).size().unstack(fill_value=0)

print("\n=== Pivot Tabel Cacah Kata (Vokal x Konsonan) ===")
print(pivot_cacah)

# --- Visualisasi heatmap cacah kata ---
plt.figure(figsize=(10,6))
sns.heatmap(pivot_cacah, annot=True, fmt="d", cmap="OrRd")
plt.title("Heatmap Cacah Kata per Kombinasi Vokal vs Konsonan")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()
```

Heatmap tersebut menggambarkan jumlah kata (cacah kata) berdasarkan kombinasi antara jumlah huruf vokal dan konsonan dalam teks yang dianalisis. Warna yang semakin gelap menunjukkan frekuensi kata yang semakin tinggi. Terlihat bahwa area dengan 2â€“3 huruf vokal dan 3â€“4 huruf konsonan memiliki intensitas warna paling pekat, menandakan bahwa sebagian besar kata dalam teks memiliki panjang sedang dengan keseimbangan antara huruf vokal dan konsonan. Sementara itu, kata yang memiliki jumlah vokal atau konsonan terlalu sedikit maupun terlalu banyak relatif jarang muncul.

```python
# Histogram jumlah kata per kalimat
plt.figure(figsize=(10,6))
plt.bar(df_kalimat_abstrak["Kalimat_ke"], df_kalimat_abstrak["Jumlah_Kata"], color="skyblue", edgecolor="black")
plt.title("Jumlah Kata pada Setiap Kalimat")
plt.xlabel("Kalimat ke-")
plt.ylabel("Jumlah Kata")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

Histogram menunjukkan distribusi jumlah kata pada setiap kalimat dalam paragraf.

```python
from collections import Counter

# --- Bersihkan dan pisahkan kata ---
kata_bersih = re.findall(r'\b\w+\b', kesimpulan.lower())

# Hitung frekuensi kata
frekuensi = Counter(kata_bersih)

# Ubah ke DataFrame
df_freq = pd.DataFrame(frekuensi.items(), columns=["Kata", "Frekuensi"]).sort_values(by="Frekuensi", ascending=False)

# Top 5 dan Bottom 5
top5 = df_freq.head(5)
bottom5 = df_freq.tail(5)

# --- Visualisasi ---
plt.figure(figsize=(12,5))

# Top 5 kata
plt.subplot(1, 2, 1)
plt.bar(top5["Kata"], top5["Frekuensi"], color="skyblue")
plt.title("Top 5 Kata Paling Sering Muncul")
plt.xlabel("Kata")
plt.ylabel("Frekuensi")
plt.xticks(rotation=45)

# Bottom 5 kata
plt.subplot(1, 2, 2)
plt.bar(bottom5["Kata"], bottom5["Frekuensi"], color="salmon")
plt.title("Bottom 5 Kata Paling Jarang Muncul")
plt.xlabel("Kata")
plt.ylabel("Frekuensi")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
